# Fact Table

#### - Select specific columns
#### - Group by movie
#### - Mean ticket sales

```python
import pandas as pd
```

```python
df = pd.read_parquet('clean_cinema.parquet')
```

```python
df.head()
```

```python
fact_cinema = df[['film_code', 'total_sales', 'tickets_sold', 'ticket_price']] \
    .groupby('film_code') \
    .agg({
        'total_sales': 'sum',
        'tickets_sold': 'sum',
        'ticket_price': 'mean'})

fact_cinema['calc_ticket_price'] = fact_cinema['total_sales'] / fact_cinema['tickets_sold']

fact_cinema['price_difference'] = ((fact_cinema['ticket_price'] / fact_cinema['calc_ticket_price']) - 1) * 100

fact_cinema.head()
```

```python
fact_cinema.to_parquet('fact_cinema.parquet')
```
